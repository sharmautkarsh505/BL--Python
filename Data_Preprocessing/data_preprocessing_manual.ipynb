```python
"""
Author:Utkarsh Sharma(sharmautkarsh2396@gmail.com)
Date: 2021-08-03 07:17:00
Last Modified by:Utkarsh Sharma(sharmautkarsh2396@gmail.com)
Last Modified time:2021-08-03 08:59:00
Title :1. Apply following steps to dataset:Handling missing data, Handling categorical data,Split the dataset into training set and test set
,Feature scaling
"""
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

data=pd.read_csv('data_preprocessing.csv')
display(data)
data.info()
print(data.dtypes)
```

```python
#fill the NaN values with the mean or median
#you could also drop the rows in certain cases, but nod a good idea in this case as dataset is anyways very small
#simpleimputer can also be used to fill missing values 

data['Age'].fillna(value=data['Age'].mean(),inplace=True)
data['Salary'].fillna(value=data['Salary'].mean(),inplace=True)
#OR
#data['Age'].fillna(value=data['Age'].median(),inplace=True)
#data['Salary'].fillna(value=data['Salary'].median(),inplace=True)

display(data)
```

```python
#to ensure no issues in encoding, coverting the target variable to category dtype
data['Purchased']=data['Purchased'].astype('category')
```

```python
#encoding the country names
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
data=ct.fit_transform(data)
```

```python
data=pd.DataFrame(data,columns=['France','Germany','Spain','Age','Salary','Purchased'])
```

```python
#splitting dset
from sklearn.utils import shuffle
data=shuffle(data)

training=data.iloc[:-2,:]
testing=data.iloc[7:9,:]

print(training.shape)
print(testing.shape)
X_train=training.iloc[:,:-1]
y_train=training.iloc[:,-1]

print(X_train.shape)
print(y_train.shape)

X_test=testing.iloc[:,:-1]
y_test=testing.iloc[:,-1]

print(X_test.shape)
print(y_test.shape)
```

```python
#scaling features age and salary
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train.iloc[:,3:]=scaler.fit_transform(X_train.iloc[:,3:])
X_test.iloc[:,3:]=scaler.transform(X_test.iloc[:,3:])

print(X_train)
```
