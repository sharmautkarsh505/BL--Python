```python
"""
Author:Utkarsh Sharma(sharmautkarsh2396@gmail.com)
Date: 2021-08-03 05:08:00
Last Modified by:----
Last Modified time:----
Title :Write a program to draw a scatter plot for random 1000 x and y coordinates
"""
import numpy as np 
import plotly.express as px

x=[np.random.randint(1,100) for _ in range(1,1001)]
y=[np.random.randint(1,100) for _ in range(1,1001)]
```

```python
fig=px.scatter(x=x,y=y)
fig.show()
```
