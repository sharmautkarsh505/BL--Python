```python
"""
Author:Utkarsh Sharma(sharmautkarsh2396@gmail.com)
Date: 2021-09-02 11:29:00
Last Modified by:----
Last Modified time:----
Title : Write a Python programming to create a pie chart with a title of the popularity of programming
Languages.
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

languages=np.array(['Java','Python', 'PHP', 'JavaScript', 'C#', 'C++'])
popularity_pts=np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
```

```python
fig=plt.figure(figsize=(12,4))
ax=fig.add_axes([0,0,1,1])

ax.pie(x=popularity_pts,labels=languages,radius=1.2,textprops=dict(color='w',fontsize=12),autopct='%1.1f%%')
ax.set_title('Popularity of Programming Languages',fontsize=14,fontweight='bold',bbox={'facecolor':'0.7', 'pad':3})


plt.show()
```
