```python
"""
Author:Utkarsh Sharma(sharmautkarsh2396@gmail.com)
Date: 2021-09-03 12:23:00
Last Modified by:----
Last Modified time:----
Title : Write a Python program to create bar plot from a DataFrame.
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=np.array([[4,8,5,7,6],[2,3,4,2,6],[4,7,4,7,8],[2,6,4,8,6],[2,4,3,3,2]])
df=pd.DataFrame(data,columns=['a','b','c','d','e'],index=[2,4,6,8,10])
```

```python
df.plot(kind='bar',legend=False)
plt.title('Bar plot of Dataframe')
plt.xlabel('Represents each column')
plt.ylabel('Represents values in each column')
locs, labels=plt.xticks()
new_xticks=['a','b','c','d','e']
plt.xticks(locs,new_xticks, rotation=0, horizontalalignment='right')
plt.show()
```
