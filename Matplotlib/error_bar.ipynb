```python
"""
Author:Utkarsh Sharma(sharmautkarsh2396@gmail.com)
Date: 2021-09-02 04:23:00
Last Modified by:----
Last Modified time:----
Title :Write a Python program to create bar plots with error bars on the same figure.
"""
import numpy as np
import matplotlib.pyplot as plt

mean_velocity=np.array([0.2474, 0.1235, 0.1737, 0.1824])
std_velocity=mean_velocity-np.array([0.3314, 0.2278, 0.2836, 0.2645])
```

```python
fig=plt.figure(figsize=(5,5))
ax=fig.add_axes([0,0,1,1])

ax.bar(x=np.arange(4),height=mean_velocity,yerr=std_velocity)
```
