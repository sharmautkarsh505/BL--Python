```python
"""
Author:Utkarsh Sharma(sharmautkarsh2396@gmail.com)
Date: 2021-09-02 11:06:00
Last Modified by:----
Last Modified time:----
Title : Write a Python programming to display a bar chart of the popularity of programming
Languages.
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

languages=np.array(['Java','Python', 'PHP', 'JavaScript', 'C#', 'C++'])
popularity_pts=np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
```

```python
plt.pie(x=popularity_pts,labels=languages,radius=1.5,textprops=dict(color='w',fontsize=12))
plt.show()
```
