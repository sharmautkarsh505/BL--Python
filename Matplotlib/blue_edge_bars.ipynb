```python
"""
Author:Utkarsh Sharma(sharmautkarsh2396@gmail.com)
Date: 2021-09-02 01:36:00
Last Modified by:----
Last Modified time:----
Title : Write a Python programming to display a bar chart of the popularity of programming
Languages. Make blue border to each bar.
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

languages=np.array(['Java','Python', 'PHP', 'JavaScript', 'C#', 'C++'])
popularity_pts=np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
```

```python
fig=plt.figure(figsize=(5,3))
ax=fig.add_axes([0,0,1,1])

bar=ax.bar(x=languages,height=popularity_pts,edgecolor='blue',color='lightgrey')
ax.set_xlabel('Programming Languages',fontsize=12,color='black')
ax.set_ylabel('Popularity Ratings',fontsize=12,color='black')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.bar_label(bar,labels=popularity_pts,padding=0.1)
plt.show()
```
