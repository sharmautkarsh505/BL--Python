```python
"""
Author:Utkarsh Sharma(sharmautkarsh2396@gmail.com)
Date: 2021-09-02 18:44:00
Last Modified by:----
Last Modified time:----
Title : Write a Python program to draw a scatter plot comparing two subject marks of Mathematics
and Science. Use marks of 10 students.
"""
import matplotlib.pyplot as plt

math_marks = [88, 92, 80, 89, 100, 80, 60, 100, 80, 34]
science_marks = [35, 79, 79, 48, 100, 88, 32, 45, 20, 30]
marks_range = ['S1','S2','S3','S4','S5','S6','S7','S8','S9','S10']
```

```python
fig=plt.figure(figsize=(8,4))
ax=fig.add_axes([0,0,1,1])

ax.scatter(x=marks_range,y=math_marks,c='steelblue')
ax.scatter(x=marks_range,y=science_marks,c='orange')
ax.legend(labels=['Maths','Science'],loc=1)

ax.set_xlabel('Student',fontsize=10,fontweight='bold')
ax.set_ylabel('Student Score in Mathematics and Science\n(out of 100)',fontsize=10,fontweight='bold')
ax.set_title('Comparison of Marks in Mathematics and Science',fontsize=12,fontweight='bold')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

```
