```python
"""
Author:Utkarsh Sharma(sharmautkarsh2396@gmail.com)
Date: 2021-09-02 04:23:00
Last Modified by:----
Last Modified time:----
Title : Write a Python program to create bar plot of scores by group and gender. Use multiple X
values on the same chart for men and women.
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
```

```python
x=np.arange(1,len(labels)+1)
print(x)
width=0.35
fig=plt.figure(figsize=(6,5))
ax=fig.add_axes([0,0,1,1])
rects1=ax.bar(x=x-width/2,height=men_means,width=0.35,color='steelblue',edgecolor='black')
rects2=ax.bar(x=x+width/2,height=women_means,width=0.35,color='grey',edgecolor='black')

#ax.legend()
#give labels to x and y axis, and title
ax.set_xlabel('Group of Men and Women',fontsize=12,fontweight='bold')
ax.set_ylabel('Scores',fontsize=12,fontweight='bold')
ax.set_title('Scores by Group and Gender',fontsize=16,fontweight='bold')

#set ticks, then label then with correct labels
ax.set_xticks(x)
ax.set_xticklabels(labels)

#label bar with the val of means
ax.bar_label(rects1,labels=men_means,padding=0.1)
ax.bar_label(rects2,labels=women_means,padding=0.1)


plt.show()
```
