```python
"""
Author:Utkarsh Sharma(sharmautkarsh2396@gmail.com)
Date: 2021-09-02 17:53:00
Last Modified by:----
Last Modified time:----
Title : Write a Python program to draw a scatter graph taking a random distribution in X and Y and
plotted against each other.
"""
import numpy as np
import matplotlib.pyplot as plt

x=[np.random.randint(1,101) for _ in range(1,101)]
y=[np.random.randint(1,101) for _ in range(1,101)]
```

```python
fig=plt.figure(figsize=(8,4))
ax=fig.add_axes([0,0,1,1])

ax.scatter(x=x,y=y,c='grey')

ax.set_xlabel('X',fontsize=12,fontweight='bold')
ax.set_ylabel('Y',fontsize=12,fontweight='bold')
ax.set_title('Scatterplot Y vs X',fontsize=16,fontweight='bold')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

```
