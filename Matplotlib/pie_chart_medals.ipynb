```python
"""
Author:Utkarsh Sharma(sharmautkarsh2396@gmail.com)
Date: 2021-09-02 11:59:00
Last Modified by:----
Last Modified time:----
Title : Write a Python programming to create a pie chart of gold medal achievements of five most
successful countries in 2016 Summer Olympics. Read the data from a csv file.
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('medal.csv')
display(data)
```

```python
#autopct issue
plt.pie(x=data['gold_medal'],explode=[0.2,0,0,0,0],labels=data['country'],textprops=dict(color='w',fontsize=12),wedgeprops=dict(edgecolor='black'),autopct='%1.1f%%')
       
```
