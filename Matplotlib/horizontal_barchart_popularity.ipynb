```python
"""
Author:Utkarsh Sharma(sharmautkarsh2396@gmail.com)
Date: 2021-09-02 00:36:00
Last Modified by:----
Last Modified time:----
Title : Write a Python programming to display a horizontal bar chart of the popularity of
programming Languages.
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

languages=np.array(['Java','Python', 'PHP', 'JavaScript', 'C#', 'C++'])
popularity_pts=np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
```

```python
fig=plt.figure(figsize=(8,4))
ax=fig.add_axes([0,0,1,1])

ax.barh(y=languages,width=popularity_pts,edgecolor='black',color='grey')
ax.set_xlabel('Popularity Ratings',fontsize=12,color='black')
ax.set_ylabel('Programming Languages',fontsize=12,color='black')
ax.set_title('Popularity of the programming languages')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```
